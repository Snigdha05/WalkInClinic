Quarterly analysis of Appointments booked by Female patients with age greater than 25 years old suffering from same primary symptoms.From this report Manager will get to know the appointmnet details booked by females suffereing from same allergies.
This will help the managerin determining the health conditions and allergies within the area/city that people are mostly visiting to same department.

This will also help to allocate and increase the number of doctor and staff availability in the specific department.

```python
import pandas as pd
import sqlite3
from datetime import datetime, date
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Sqlconnection = sqlite3.connect("Walk-In-Clinic-DB.db")
```

```python
Patientdf = pd.read_sql("SELECT * FROM PatientData", con=Sqlconnection)



```

```python
Appointmentdetailsdf = pd.read_sql("select * from PatientData as pat inner join PatientAppointment as app on Pat.PatientNumber=app.PatientNumber ", con=Sqlconnection)
```

```python
# Extracting Year and Quarter from Date feild

Appointmentdetailsdf['Year']=pd.to_datetime(Appointmentdetailsdf['ScheduledDate'], format='%Y-%m-%d').dt.year


Appointmentdetailsdf['Quarter'] = pd.to_datetime(Appointmentdetailsdf['ScheduledDate'], format='%Y-%m-%d').dt.to_period('Q')

Appointmentdetailsdf["FullName"] = Appointmentdetailsdf["FirstName"] +" " +Appointmentdetailsdf["LastName"]

```

```python
######## Calculating Age from DOB column
```

```python
def age(born):
    birthdate= datetime.strptime(born, "%Y-%m-%d").date()
    currentdate = date.today()
    return currentdate.year - birthdate.year - ((currentdate.month,  currentdate.day) < (birthdate.month,  birthdate.day))
  
Appointmentdetailsdf['Age'] = Appointmentdetailsdf['DOB'].apply(age)
  
display(Appointmentdetailsdf)
```

```python
def select_columns(data_frame, column_names):
    new_frame = data_frame.loc[:, column_names]
    return new_frame

```

```python

selected_columns = ['Year', 'Quarter','PrimarySymptoms','PatientNumbers', 'FullName',"Gender","Age","EmailID","DoctorName","AppointmentID","ScheduledDate","ScheduledTime","ConfirmedFlag"]
Appointmentdf = select_columns(Appointmentdetailsdf, selected_columns)


```

```python
####### Filter the data based on Gender='F' and Age>='25'
```

```python
Agefilter=Appointmentdf['Age'] >=25
Agefilter
```

```python
ConfirmedFlagfilter=Appointmentdf['ConfirmedFlag'] ==1
ConfirmedFlagfilter
```

```python
Appointmentdf.where(Femalefilter & Agefilter & ConfirmedFlagfilter, inplace = True)
```
